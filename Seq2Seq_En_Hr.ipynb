#BLEU

```python
!pip install evaluate
import evaluate
```

```python
hr_ref = open('Data/hr_tgt-test.tokenized.txt', encoding="utf-8")
```

```python
hr_pred = open('Data/hr_to_en_pred_1000.txt', encoding="utf-8")
```

```python
references = []
predictions = []
```

```python
for line in hr_ref:
  references.append(str(line))

references_new = []
for line in references:
  line = line.replace("\n","")
  references_new.append(line)
```

```python
for line in hr_pred:
  predictions.append(str(line))

predictions_new = []
for line in predictions:
  line = line.replace("\n","")
  predictions_new.append(line)
```

```python
bleu = evaluate.load("bleu")
```

```python
results = bleu.compute(predictions=predictions_new, references=references_new)
```

```python
results
```

```python
gbleu = evaluate.load("google_bleu")
```

```python
results = gbleu.compute(predictions=predictions_new, references=references_new)
```

```python
results
```
